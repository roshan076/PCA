# Random Gaussian distribution

```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate a normal(Gaussian) distribution
np.random.seed(0)
mat1 = np.random.randn(20,2)
```

```python
#Plot the generated matrix
plt.scatter(mat1[:,0], mat1[:,1], marker='d')
plt.suptitle("Original Data", fontsize=12, fontweight='bold')
plt.title("(THA076BEI005 & THA076BEI025)", fontsize=10)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
#Generate uniform distribution matrix between[0,1]
mat2 = np.random.uniform(0, 1, (2,2))
```

```python
#Matrix multiplication
mat_product = np.dot(mat1,mat2) 
```

```python
#Plot the product matrix
plt.scatter(mat_product[:,0],mat_product[:,1], marker='d')
plt.suptitle("Transformed Data", fontsize=12, fontweight='bold')
plt.title("(THA076BEI005 & THA076BEI025)", fontsize=10)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
#Calculate variance against both axis
var_x = np.var(mat_product[:,0])
var_y = np.var(mat_product[:,1])
print("Variance along x-axis: ", var_x)
print("Variance along x-axis: ", var_y)
```

```python
#Calculate the Covariance matrix
Sx = (1/(len(mat_product)-1))*np.dot(mat_product.T,mat_product)
print("Covariance Matrix\n",Sx)
```

```python
#Calculate eigenvalues & eigenvectors
eigenvalues, eigenvector = np.linalg.eig(Sx)
print("Eigen values\n",eigenvalues)
print("\nEigen vector\n",eigenvector)
```

```python
#Variance of eigenvalues
total_var = np.sum(eigenvalues)
var_prop = (eigenvalues / total_var) * 100
print("Variance proportion: ",var_prop)
```

```python
#Compute output
y = np.dot(eigenvector.T,mat_product.T)
y.shape
```

```python
#Covaraince matrix of output
Sy = (1/(len(y)-1))*np.dot(y,y.T)
print("Covaraince matrix\n",Sy)
```

```python
#Plot y uisng both eigenvectors
plt.scatter(y[0,:],y[1,:], marker='d')
plt.suptitle("PCA Data", fontsize=12, fontweight='bold')
plt.title("(THA076BEI005 & THA076BEI025)", fontsize=10)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.axis('equal')
plt.grid('True')
plt.show()
```

```python
#Plot y uisng one eigenvector
y_1d = np.dot(eigenvector.T[:,0],mat_product.T)

plt.scatter(y_1d, np.zeros_like(y_1d), marker='d')
plt.suptitle("PCA Data (1D)", fontsize=12, fontweight='bold')
plt.title("(THA076BEI005 & THA076BEI025)", fontsize=10)
plt.xlabel("PC1")
plt.axis('equal')
plt.grid(True)
plt.show()
```
